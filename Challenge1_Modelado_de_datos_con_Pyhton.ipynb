### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_t1 = tienda['Precio'].sum()
facturacion_t2 = tienda2['Precio'].sum()
facturacion_t3 = tienda3['Precio'].sum()
facturacion_t4 = tienda4['Precio'].sum()

print(f"Facturación Tienda 1: ${facturacion_t1:,.0f}")
print(f"Facturación Tienda 2: ${facturacion_t2:,.0f}")
print(f"Facturación Tienda 3: ${facturacion_t3:,.0f}")
print(f"Facturación Tienda 4: ${facturacion_t4:,.0f}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()

print("Ventas por Categoría - Tienda 1:")
display(ventas_por_categoria_tienda1)

print("\nVentas por Categoría - Tienda 2:")
display(ventas_por_categoria_tienda2)

print("\nVentas por Categoría - Tienda 3:")
display(ventas_por_categoria_tienda3)

print("\nVentas por Categoría - Tienda 4:")
display(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
productos_vendidos_tienda1 = tienda['Producto'].value_counts()
productos_vendidos_tienda2 = tienda2['Producto'].value_counts()
productos_vendidos_tienda3 = tienda3['Producto'].value_counts()
productos_vendidos_tienda4 = tienda4['Producto'].value_counts()

print("Productos más vendidos Tienda 1:")
print(productos_vendidos_tienda1.head())
print("\nProductos menos vendidos Tienda 1:")
print(productos_vendidos_tienda1.tail())

print("\nProductos más vendidos Tienda 2:")
print(productos_vendidos_tienda2.head())
print("\nProductos menos vendidos Tienda 2:")
print(productos_vendidos_tienda2.tail())

print("\nProductos más vendidos Tienda 3:")
print(productos_vendidos_tienda3.head())
print("\nProductos menos vendidos Tienda 3:")
print(productos_vendidos_tienda3.tail())

print("\nProductos más vendidos Tienda 4:")
print(productos_vendidos_tienda4.head())
print("\nProductos menos vendidos Tienda 4:")
print(productos_vendidos_tienda4.tail())
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio Tienda 1: {costo_envio_promedio_tienda1:.2f}")
print(f"Costo de envío promedio Tienda 2: {costo_envio_promedio_tienda2:.2f}")
print(f"Costo de envío promedio Tienda 3: {costo_envio_promedio_tienda3:.2f}")
print(f"Costo de envío promedio Tienda 4: {costo_envio_promedio_tienda4:.2f}")
```

# 6. Visualizaciones

```python
import matplotlib.pyplot as plt
import numpy as np

# Data de Facturación
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
revenues = [facturacion_t1, facturacion_t2, facturacion_t3, facturacion_t4]

# Gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(stores, revenues, color=plt.cm.Paired.colors)
plt.ylabel('Ingreso Total')
plt.title('Ingreso Total por Tienda')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2,
             height + 300,
             f'${height:,.0f}',
             ha='center', va='bottom')
plt.show()

# Data de calificación promedio
ratings = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]

# Gráfico de barras horizontales

plt.figure(figsize=(10, 6))
bars = plt.barh(stores, ratings,color=plt.cm.Paired.colors)
plt.xlabel('Calificación Promedio')
plt.title('Calificación Promedio por Tienda')
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.05,
             bar.get_y() + bar.get_height() / 2,
             f'{width:.2f}',
             va='center')

plt.tight_layout()
plt.show()

# Data de ventas por categoría en cada tienda
stores_data = {
    'Tienda 1': tienda['Categoría del Producto'].value_counts(),
    'Tienda 2': tienda2['Categoría del Producto'].value_counts(),
    'Tienda 3': tienda3['Categoría del Producto'].value_counts(),
    'Tienda 4': tienda4['Categoría del Producto'].value_counts(),
}


# Grid 2x2
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()

colors = plt.cm.Paired.colors

#Gráficos de pie

for i, (store_name, sales_data) in enumerate(stores_data.items()):
    wedges, texts, autotexts = axs[i].pie(
        sales_data,
        labels=None,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors
    )
    axs[i].set_title(f'Distribución de Ventas por Categoría en {store_name}')
    axs[i].axis('equal')

    # Añadir la leyenda
    axs[i].legend(wedges, sales_data.index,
                  title="Categorías",
                  loc="center left",
                  bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```

## Informe Final: Recomendación de Tienda para el Sr. Juan

### Introducción

El propósito de este informe es analizar el rendimiento de las cuatro tiendas proporcionadas y, con base en los hallazgos, recomendar al Sr. Juan cuál de ellas sería la mejor opción para vender. Se han evaluado métricas clave como ingresos totales, ventas por categoría, calificación promedio de los clientes, productos más y menos vendidos, y costos de envío promedio.

### Análisis de Datos y Visualizaciones

**Ingresos Totales:**

Como se muestra en el gráfico de barras de "Ingreso Total por Tienda", la Tienda 1 generó los ingresos totales más altos, seguida por la Tienda 2, la Tienda 3 y la Tienda 4. Esto sugiere que la Tienda 1 tiene el mayor volumen de ventas en términos monetarios.

**Ventas por Categoría:**

El análisis de ventas por categoría (ejemplificado con el gráfico de pastel para la Tienda 1) revela la distribución de las ventas entre las diferentes categorías de productos en cada tienda. Observamos que las categorías de "Muebles" y "Electrónicos" son consistentemente las más vendidas en todas las tiendas. Comprender estas distribuciones es crucial para identificar las fortalezas de cada tienda en términos de inventario y demanda.

**Calificación Promedio de los Clientes:**

El gráfico de línea de "Calificación Promedio por Tienda" muestra que las calificaciones promedio de los clientes son relativamente similares entre las tiendas, todas por encima de 3.9. La Tienda 3 tiene la calificación promedio ligeramente más alta, lo que podría indicar una mayor satisfacción del cliente en esa tienda.

**Productos Más y Menos Vendidos:**

El análisis de los productos más y menos vendidos en cada tienda proporciona información detallada sobre la popularidad de productos específicos. Identificar los productos de alto rendimiento puede ayudar a comprender qué impulsa las ventas en cada ubicación, mientras que los productos de bajo rendimiento pueden indicar áreas de mejora o productos que podrían descontinuarse.

**Costo de Envío Promedio:**

El costo de envío promedio varía entre las tiendas, siendo la Tienda 4 la que tiene el costo de envío promedio más bajo. Esto podría ser un factor a considerar en la rentabilidad, aunque los ingresos totales y el volumen de ventas son generalmente indicadores más significativos del rendimiento general.

### Conclusión y Recomendación

Considerando todos los factores analizados, **se recomienda que el Sr. Juan venda la Tienda 1**.

Aunque las calificaciones promedio son similares entre las tiendas y la Tienda 4 tiene el costo de envío promedio más bajo, la Tienda 1 destaca significativamente en términos de **ingresos totales**. Un mayor ingreso total sugiere una base de clientes más sólida y un mayor volumen general de transacciones, lo que la convierte en una opción más atractiva para la venta. Además, el análisis de ventas por categoría y productos más vendidos en la Tienda 1 puede proporcionar información valiosa sobre qué productos son más exitosos en esa ubicación, lo que podría ser beneficioso para un nuevo propietario.

Si bien las otras tiendas tienen sus puntos fuertes (como la calificación promedio de la Tienda 3 o el bajo costo de envío de la Tienda 4), el desempeño financiero superior de la Tienda 1 la posiciona como la opción más rentable y con mayor potencial para el Sr. Juan.
